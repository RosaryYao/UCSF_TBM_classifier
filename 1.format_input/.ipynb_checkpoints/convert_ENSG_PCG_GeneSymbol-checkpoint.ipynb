```python
import pandas as pd
```

```python
df1 = pd.read_csv('ENSG_combined.txt', sep = '\t')
df2 = pd.read_csv('ENSG_GeneSymbol_list.csv')
```

```python
df1.shape, df2.shape
```

```python
df3 = df1.merge(df2, how='inner', left_on='geneID', right_on='original')
```

```python
df4 = df3[df3['gene_biotype'] == 'protein_coding']
```

```python
df4.reset_index().drop('index',axis=1).drop('geneID',axis=1).set_index('ensembl').shape
```

```python
test_set = df4.reset_index().drop('index',axis=1).drop('geneID',axis=1).set_index('ensembl').iloc[:,:-7]
```

```python
test_set.to_csv('final_test_set.csv')
```
